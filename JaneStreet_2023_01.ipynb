```julia
function step(a::Int, b::Int, c::Int, d::Int)
    return (abs(a-b), abs(b-c), abs(c-d), abs(d-a))
end

function f(z::Int, a::Int, b::Int, c::Int, print=false)
    s = 1
    while a!=0 || b!=0 || c!=0 || z!=0
        if print
            if (z > 0) & (a > 0) & (b > 0) & (c > 0)
                az, ba, cb, = a/z, b/a, c/b
                println("($z, $a, $b, $c) with ratios $az, $ba, $cb ")
            else
                println("($z, $a, $b, $c)")
            end
        end
        z,a,b,c = step(z, a,b,c)
        s +=1
    end
    return s
end    
```

```julia
@assert f(10,6,3,1) == 7
```

```julia
ratio = 0.54368901269207

d = 9999966
c = trunc(Int, d * ratio)
b = trunc(Int, c * ratio)
a = trunc(Int, b * ratio)


f(a, b, c, d, true)
```
